2. Investigate the data in Table 4.5 of the last upper molar of the extinct mammal

(a) Construct a frequency distribution table using appropriate intervals.
(b) Display the distribution as a histogram (make a bar chart of the table you
constructed in the previous exercise).
(c) Describe the shape of the distribution.

```python
# loading the data into the dataframe
df_um = pd.read_csv('molar.csv')
```

```python
# getting overview over the data - the ranges and frequency of the ranges

um_value_count = df_um.value_counts()

# sorting the values
um_value_count.sort_index()
```

```python
# we divide the molar data into approximately 7 intervals (I have chosen 7, since the distribution looked the best, but others might be more appropriate.
# read more about the cut function in pandas in D

df_um_cut = pd.cut(df_um['Upper Molar'], bins=7)

# displaying the sorted interval from smallest to largest
df_um_cut.value_counts().sort_index()
```

```python
# plotting the frequency distribution of the upper molars as a bar plot
df_um_cut.value_counts().sort_index().plot(kind='bar')
plt.ylabel('Frequency')
plt.xlabel('Width of upper molar (mm)')
plt.title('Width of the upper molar of Acropihecus rigidus')
```

```python
df_um.plot.hist(bins=7)
```

## Excercise 4.5 Make a dot plot
The following exercise is optional. Unfortunately, the dot plot is not implemented as one
of Pandas plotting method (some methods are listed in Table 4.8), but the matplotlib
module can be used. The code block below generate a dotplot of the infant mortality
rate data discussed in the chapter.

   1. Make a dotplot of the infant mortality rate data discussed in the chapter. See Table
    4.2 (page 51) and code block below.


```python
# DataFrame with data
df = pd.DataFrame({
    'Infant mortality rate': [
        11.4 , 44.7 , 22.6 , 7.7 , 18.9 , 20.9 ,
        30.0 , 24.7 , 28.6 , 18.8 , 11.3 , 26.5
    ]
})

# Make dot plot.
y = df['Infant mortality rate']
plt.figure(figsize=[6, 1]) # new elongated figure
plt.plot( # plot values against zeroes:
        y, # - values on x-axis
        y * 0, # - values on y-axis
        'o' # -make dots
)
plt.yticks([]) # remove ticks on y-axis.
plt.xlim(0, 50) # set limits on x-axis
plt.xlabel('Infant mortality rate in percent')
plt.show()
```
